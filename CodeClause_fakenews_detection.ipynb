```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
#Loading the fake & true datasets
fd=pd.read_csv('/content/Fake.csv')
td=pd.read_csv('/content/True.csv')
```

```python
fd.head()
```

```python
td.head()
```

```python
#classifying fakenews as 0 and true news as 1

fd["class"]=0
td["class"]=1
```

```python
fd.shape,td.shape
```

```python
#manual testing last 10 rows of data and Dropping the last 10 rows

fd_manual_testing=fd.tail(10)
for i in range(23480,23470,-1):
    fd.drop([i],axis=0,inplace=True)

td_manual_testing=td.tail(10)
for i in range(21416,21406,-1):
    td.drop([i],axis=0,inplace=True)
```

```python
fd.shape,td.shape
```

```python
#merge data

df_manual_testing=pd.concat([fd_manual_testing,td_manual_testing],axis=0)
df_manual_testing.to_csv("manual_testing.csv")
```

```python
fd_manual_testing["class"]=0
td_manual_testing["class"]=1
```

```python
fd_manual_testing.head(10)
```

```python
td_manual_testing.head(10)
```

```python
#merging two datasets

data_merge=pd.concat([fd,td],axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
#deleting unwanted columns

data=data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
#random shuffling of the data frames

data=data.sample(frac=1)
```

```python
data.head()

#Class 1 - true dataset
#class 0 - fake news
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
#to remove special characters

def wordopt(text):
    text=text.lower()
    text=re.sub('\[.*?\]','',text)
    text=re.sub("\\W"," ",text)
    text=re.sub('https?://\S+|www\.\S+', '',text)
    text=re.sub('<.*?>+', '',text)
    text=re.sub('[%s]'%re.escape(string.punctuation), '',text)
    text=re.sub('\n', '', text)
    text=re.sub('\w*\d\w*', '',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
#Removed all unwanted characters

data.head()
```

```python
x=data['text']
y=data['class']
```

```python
#splitting the data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
#Logistic regression

from sklearn.linear_model import LogisticRegression

LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
#making prediction

pred_lr=LR.predict(xv_test)
```

```python
#Accuracy

LR.score(xv_test,y_test)
```

```python
#classification report

print(classification_report(y_test,pred_lr))
```

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier

DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
#Gradient Boosting Classifier

from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
pred_gb=GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
def output_label(n):
    if n==0:
        return "FAke news"
    elif n==1:
        return "Not a Fake News"

def manual_testing(news):
    testing_news={"text":[news]}
    new_def_test=pd.DataFrame(testing_news)
    new_def_test["text"]=new_def_test["text"].apply(wordopt)
    new_x_test=new_def_test["text"]
    new_xv_test=vectorization.transform(new_x_test)
    pred_lr=LR.predict(new_xv_test)
    pred_dt=DT.predict(new_xv_test)
    pred_gb=GB.predict(new_xv_test)
    pred_rf=RF.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output_label(pred_lr[0]),
                                                                                                            output_label(pred_dt[0]),
                                                                                                            output_label(pred_gb[0]),
                                                                                                            output_label(pred_rf[0])))
```
