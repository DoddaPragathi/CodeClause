```python
#Importing the libraries

import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
from textblob import TextBlob

from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
from wordcloud import WordCloud
from nltk.sentiment.vader import SentimentIntensityAnalyzer

import cufflinks as cf
%matplotlib inline
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected =True)
cf.go_offline();
import plotly.graph_objs as go
from plotly.subplots import make_subplots

import warnings
warnings.filterwarnings("ignore")
warnings.warn("this will not show")

pd.set_option('display.max_columns',None)

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
text_df= da.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date','hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'],axis=1)
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
#Text Processing

def data_processing(text):
    text = text.lower()
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]
    return " ".join(filtered_text)
```

```python
text_df.text = text_df['text'].apply(data_processing)
```

```python
text_df = text_df.drop_duplicates('text')
```

```python
#Stemming reduces the number of unique words that need to be processed by an algorithm, which can improve its performance.

stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
```

```python
text_df['text'] = text_df['text'].apply(lambda x: stemming(x))
```

```python
text_df.head()
```

```python
print(text_df['text'].iloc[0],"\n")
print(text_df['text'].iloc[1],"\n")
print(text_df['text'].iloc[2],"\n")
print(text_df['text'].iloc[3],"\n")
print(text_df['text'].iloc[4],"\n")
```

```python
text_df.info()
```

```python
def polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
text_df['polarity'] = text_df['text'].apply(polarity)
```

```python
text_df.head(10)
```

```python
def sentiment(label):
    if label <0:
        return "Negative"
    elif label ==0:
        return "Neutral"
    elif label>0:
        return "Positive"
```

```python
text_df['sentiment'] = text_df['polarity'].apply(sentiment)
```

```python
text_df.head()
```

```python
vect = CountVectorizer(ngram_range=(1,2)).fit(text_df['text'])
```

```python
feature_names = vect.get_feature_names_out()
print("Number of features: {}\n".format(len(feature_names)))
print("First 20 features:\n {}".format(feature_names[:20]))
```

```python
X=text_df['text']
Y=text_df['sentiment']
X=vect.transform(X)
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print("size of x_train",(x_train.shape))
print("size of y_train",(y_train.shape))
print("size of x_test",(x_test.shape))
print("size of y_test",(y_test.shape))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
lr_acc=accuracy_score(lr_pred,y_test)
print("Test Accuracy: {:.2f}%".format(lr_acc*100))
```

```python
print(confusion_matrix(y_test,lr_pred))
print("\n")
print(classification_report(y_test,lr_pred))
```

```python
style.use('classic')
cm=confusion_matrix(y_test,lr_pred, labels=lr.classes_)
dis=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=lr.classes_)
dis.plot()
```

```python
#Hyper parameter tuning

from sklearn.model_selection import GridSearchCV

param_grid={'C':[0.001,0.01,0.1,1,10]}
grid=GridSearchCV(LogisticRegression(),param_grid)
grid.fit(x_train,y_train)
```

```python
print("Best Parameters:",grid.best_params_)
```

```python
y_pred=grid.predict(x_test)
```

```python
lr_acc=accuracy_score(y_pred,y_test)
print("Test Accuracy: {:.2f}%".format(lr_acc*100))
```

```python
print(confusion_matrix(y_test,y_pred))
print("\n")
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import LinearSVC

svm=LinearSVC()
svm.fit(x_train,y_train)
```

```python
svm_pred=svm.predict(x_test)
svc_acc=accuracy_score(svm_pred,y_test)
print("Test Accuracy: {:.2f}%".format(svc_acc*100))
```

```python
print(confusion_matrix(y_test,svm_pred))
print("\n")
print(classification_report(y_test,svm_pred))
```
